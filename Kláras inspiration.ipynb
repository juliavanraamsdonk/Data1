```python

# Dropping rows where all values are missing
df_cleaned = df_new.dropna(how='all')

# Replacing all remaining missing values with "unknown"
df_cleaned.fillna('unknown', inplace=True)

#Replace missing values in numerical columns with 0
numerical_cols = df_cleaned.select_dtypes(include=['float64', 'int64']).columns
df_cleaned[numerical_cols] = df_cleaned[numerical_cols].fillna(0)

# Replace missing values in non-numerical columns with "unknown"
non_numerical_cols = df_cleaned.select_dtypes(exclude=['float64', 'int64']).columns
df_cleaned[non_numerical_cols] = df_cleaned[non_numerical_cols].fillna('unknown')


print(df_cleaned)
```
