## Exercise 3: Manage Missing Data
We’ve got a number of ways in general of dealing with missing data. These involve

1. Dropping off cases (or rows) in the data with any missing variables
2. Excluding variables in the data with any missing data 
3. Selectively choosing indicators with only a limited amount of missing data
4. Replacing missing variables with averages, or other representative values
5. Creating a separate model to predict missing data

- Count the missing values in each column
- Manage the missing values (delete or replace values or leave them as they are) and briefly explain your choice for each column


```python
# Counting the number of missing values in each column
missing_values = df.isnull().sum()

print(f"Number of missing values in each column (out of {len(df)} entries):")
display(missing_values)

# They are: rating, isbn, year, genre_urls
for col in ['rating', 'isbn', 'year', 'genre_urls']:
    print(f"Rows with missing values in '{col}' column:")
    display(df[df[col].isnull()])

# Drop rows with missing rating (since useless rows with lots of missing values)
df = df.dropna(subset=['rating'])

# As for the rest, you could remove the variables (finding representative values is not easily done with this dataset), but since we're not sure what to do with the data, removing the rows is the best option

df = df.dropna()
```

## Exercise 4: Shape the data
- Parse the `author_url` to create new column named `author`
- Sort the data by putting higher rates go first. If there are overlapping rates, try to put earlier years go first.
- **(Stretch Goal)** Examine how many books were published at each year and find lowest, highest rate of each year. 

```python
print(df['author_url'].loc[123])

# Parse author_url to get new column 'author'
df['author'] = df['author_url'].str.split('.').str.get(-1)

print("New column 'author' after parsing 'author_url':")
display(df['author'])

# Sort dataframe by rating
df = df.sort_values('rating', ascending=False)

# Find lowest and highest rate of each year and print author, name of the book and rating
lowest_rating = df.groupby('year').last()   # last() is used to get the lowest rating because the dataframe is sorted in descending order
highest_rating = df.groupby('year').first()

print("Lowest rating of each year:")
display(lowest_rating[['author', 'name', 'rating']])

print("Highest rating of each year:")
display(highest_rating[['author', 'name', 'rating']])
```

## Exercise 6: Investigate the relationship between the number of reviews and the average rating for books in the dataset cleaned-goodreads.csv procided.

- Calculate the correlation coefficient. Give me a short definition of this coefficient
- Create a scatter plot showing the relationship between these two features.
- Based on the plot and the correlation, provide a brief interpretation of the relationship.

### Python Tools: Use pandas and numpy for correlation, and matplotlib or seaborn for the scatter plot.

```python
# Relation between number of reviews and rating
df = pd.read_csv('data\cleaned-goodreads.csv')

# Calculate correlation coefficient
correlation = df['rating'].corr(df['review_count'])

print(f"Correlation coefficient between rating and review count: {correlation}")

# A correlation coeffecient of -0.038 indicates a very weak negative correlation between rating and review count. This means that (based on this data) having more reviews means a slightly lower rating, but the correlation is so weak that it is not significant.

# Scatter plot of rating and review count
import matplotlib.pyplot as plt

plt.scatter(df['rating'], df['review_count'])
plt.xlabel('Rating')
plt.ylabel('Review Count')
plt.title('Rating vs Review Count')
plt.show()

# The scatter plot shows that there is no clear relationship between rating and review count. They seem bunched up around a low review count (0 - 20000) and around a rating of 3.5 - 4.5, but no trend can be seen. This confirms the weak correlation coefficient.
```

## Exercise 7: Calculate the following descriptive statistics for the numerical features (e.g., number of reviews, average rating, etc.):
- Mean
- Median
- Standard Deviation
- Range
- Create a histogram or box plot for at least one of the numerical features, highlighting any skewness or outliers.
    
### Python Tools: Use pandas for data manipulation and matplotlib or seaborn for visualization.

```python
# For the numerical features we calculate: mean, median, standard deviation, range. We also create a histogram for one feature.

numerical_features = ['rating', 'review_count', 'year', 'rating_count'] # Skipped 'isbn' because it is essentially a name / ID.

for feature in numerical_features:
    print(f"Feature: {feature}")
    print(f"Mean: {df[feature].mean()}")
    print(f"Median: {df[feature].median()}")
    print(f"Standard deviation: {df[feature].std()}")
    print(f"Range: {df[feature].max() - df[feature].min()}")
    print()

# Histogram for 'rating'
plt.hist(df['rating'], bins=20)
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Histogram of Rating')
plt.show()

# Note the small bump at the 2.0 rating. A single book has a rating of 2.0, which is the lowest rating in the dataset. This could be seen as an outlier but is still valid data.

# find books with rating of 2.0
rating_2 = df[df['rating'] == 2]

print("Books with rating of 2.0:")
display(rating_2[['author', 'name', 'rating']])
```
