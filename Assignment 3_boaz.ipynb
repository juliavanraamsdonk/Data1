## Exercise 4: Shape the data
- Parse the `author_url` to create new column named `author`
- Sort the data by putting higher rates go first. If there are overlapping rates, try to put earlier years go first.
- **(Stretch Goal)** Examine how many books were published at each year and find lowest, highest rate of each year. 

```python
#print(np.sum(db.year.isnull()))
#not important

db['year'] = pd.to_numeric(db['year'], errors='coerce')
#skip values 'Unknows' to make column numeric

#change data types from strings to integers/float
db.rating=db.rating.astype(int)
db.rating_count=db.rating_count.astype(int)
db.review_count=db.review_count.astype(int)
#db.year=db.year.astype(int, errors='coerce')
#how do I change the datatype from float to int when there are values 'Unknown'
db.dtypes
```

```python
test_string = db.author_url[0]
test_string

#split author name from url
test_string.split('/')[-1].split('.')[1:][0]
#how this works ('https://www.goodreads.com/author/show/153394.Suzanne_Collins' is the url): I first split at the last '/', I indicate the last with [-1]
#then I split again starting at the dot where I use [1:] which extracts all elements in the list starting form index 1 onwards
#[0] grabs the first element of the list 

# A function that accepts an author url and returns the author's name
def get_author(url):

    name = url.split('/')[-1].split('.')[1:][0]
     
    return name

#Apply the get_author function to the 'author_url' column using '.map' 
#and add a new column 'author' to store the names
db['author'] = db.author_url.map(get_author)
db.author[0:5]
```

```python
#sort rating with higher rates first
rating_sorted = db.sort_values(by=['rating', 'year'], ascending=False)
rating_sorted
```

```python
#group data based on all columns of year?
#if more than one row exists with a certain value these are grouped
dbgb_year = db.groupby('year')
type(dbgb_year)

dbgb_year.count()
```
